## Setup Hand Model

```python
# Import necessary modules
import numpy as np
from hand_mechanics import Hand, visualize_hand

# Initialize the hand model with default parameters
hand = Hand()

# Alternatively, initialize with custom parameters
# hand = Hand(
#     finger_spacing=[-2, -1, 0, 1],  # Custom spacing along X-axis
#     finger_lengths={
#         'index': {'mcp_to_pip': 3.2, 'pip_to_dip': 2.1, 'dip_to_tip': 1.6},
#         'middle': {'mcp_to_pip': 3.7, 'pip_to_dip': 2.6, 'dip_to_tip': 1.6},
#         'ring': {'mcp_to_pip': 3.1, 'pip_to_dip': 2.1, 'dip_to_tip': 1.6},
#         'pinky': {'mcp_to_pip': 2.6, 'pip_to_dip': 1.6, 'dip_to_tip': 1.1},
#     },
#     thumb_length={'cmc_to_mcp': 0.7, 'mcp_to_ip': 1.0, 'ip_to_tip': 0.8}
# )

# Set joint angles
joint_angles = {
    # Finger joint angles (list of angles in degrees)
    'index_mcp': [10, 5],  # [abduction, flexion]
    'index_pip': [20],
    'index_dip': [15],
    'middle_mcp': [15, 0],
    'middle_pip': [30],
    'middle_dip': [20],
    'ring_mcp': [10, -5],
    'ring_pip': [25],
    'ring_dip': [15],
    'pinky_mcp': [5, -10],
    'pinky_pip': [15],
    'pinky_dip': [10],
    # Thumb joint angles
    'thumb_cmc': [-30, 0],  # [abduction, flexion]
    'thumb_mcp': [30],
    'thumb_ip': [15]
}

# Set the joint angles in the hand
hand.set_joint_angles(joint_angles)

# Compute forward kinematics
hand.compute_forward_kinematics()

# Visualize the hand model with Plotly
visualize_hand(hand, use_plotly=True)
```
